```python

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
#1. LOAD THE DATASET
# ============================================
print("=" * 60)
print("TITANIC DATASET - EXPLORATORY DATA ANALYSIS")
print("=" * 60)

# Load train and test datasets
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

print(f"\nTrain Dataset Shape: {train_df.shape}")
print(f"Test Dataset Shape: {test_df.shape}")
```

```python
# 2. BASIC INFORMATION
# ============================================
print("\n" + "=" * 60)
print("DATASET INFO")
print("=" * 60)
print("\nFirst 5 rows:")
print(train_df.head())

print("\nDataset Info:")
train_df.info()

print("\nColumn Names:")
print(train_df.columns.tolist())
```

```python
# 3. STATISTICAL SUMMARY
# ============================================
print("\n" + "=" * 60)
print("STATISTICAL SUMMARY")
print("=" * 60)
print(train_df.describe())

print("\nDescriptive Statistics (All columns):")
print(train_df.describe(include='all'))
```

```python
#4. MISSING VALUES ANALYSIS
# ============================================
print("\n" + "=" * 60)
print("MISSING VALUES ANALYSIS")
print("=" * 60)

missing_values = train_df.isnull().sum()
missing_percentage = (missing_values / len(train_df)) * 100
missing_df = pd.DataFrame({
    'Missing_Count': missing_values,
    'Percentage': missing_percentage
})
missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)
print(missing_df)
```

```python
# Visualize missing values
plt.figure(figsize=(10, 6))
missing_df['Percentage'].plot(kind='barh', color='coral')
plt.xlabel('Percentage of Missing Values')
plt.title('Missing Values in Titanic Dataset')
plt.tight_layout()
plt.show()
```

```python
#5. VALUE COUNTS FOR CATEGORICAL VARIABLES
# ============================================
print("\n" + "=" * 60)
print("CATEGORICAL VARIABLES ANALYSIS")
print("=" * 60)

categorical_cols = ['Survived', 'Pclass', 'Sex', 'Embarked']

for col in categorical_cols:
    if col in train_df.columns:
        print(f"\n{col} Value Counts:")
        print(train_df[col].value_counts())
        print(f"Percentage Distribution:")
        print(train_df[col].value_counts(normalize=True) * 100)
```

```python
# 6. UNIVARIATE ANALYSIS
# ============================================
print("\n" + "=" * 60)
print("UNIVARIATE ANALYSIS")
print("=" * 60)

# Survival Rate
plt.figure(figsize=(14, 5))

plt.subplot(1, 3, 1)
survival_counts = train_df['Survived'].value_counts()
plt.pie(survival_counts, labels=['Died', 'Survived'], autopct='%1.1f%%', 
        colors=['#ff6b6b', '#51cf66'], startangle=90)
plt.title('Overall Survival Rate')

# Age Distribution
plt.subplot(1, 3, 2)
train_df['Age'].hist(bins=30, edgecolor='black', color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

# Fare Distribution
plt.subplot(1, 3, 3)
train_df['Fare'].hist(bins=30, edgecolor='black', color='lightcoral')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Distribution')

plt.tight_layout()
plt.show()

# Passenger Class Distribution
plt.figure(figsize=(14, 5))

plt.subplot(1, 3, 1)
sns.countplot(data=train_df, x='Pclass', palette='viridis')
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')

plt.subplot(1, 3, 2)
sns.countplot(data=train_df, x='Sex', palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')

plt.subplot(1, 3, 3)
sns.countplot(data=train_df, x='Embarked', palette='Set1')
plt.title('Embarkation Port Distribution')
plt.xlabel('Port')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# 7. BIVARIATE ANALYSIS
# ============================================
print("\n" + "=" * 60)
print("BIVARIATE ANALYSIS")
print("=" * 60)

# Survival by Gender
plt.figure(figsize=(14, 5))

plt.subplot(1, 3, 1)
sns.countplot(data=train_df, x='Sex', hue='Survived', palette='coolwarm')
plt.title('Survival by Gender')
plt.legend(title='Survived', labels=['No', 'Yes'])

# Survival by Class
plt.subplot(1, 3, 2)
sns.countplot(data=train_df, x='Pclass', hue='Survived', palette='coolwarm')
plt.title('Survival by Passenger Class')
plt.legend(title='Survived', labels=['No', 'Yes'])

# Survival by Embarkation Port
plt.subplot(1, 3, 3)
sns.countplot(data=train_df, x='Embarked', hue='Survived', palette='coolwarm')
plt.title('Survival by Embarkation Port')
plt.legend(title='Survived', labels=['No', 'Yes'])

plt.tight_layout()
plt.show()

# Survival Rate Analysis
print("\nSurvival Rate by Gender:")
print(train_df.groupby('Sex')['Survived'].mean() * 100)

print("\nSurvival Rate by Class:")
print(train_df.groupby('Pclass')['Survived'].mean() * 100)

print("\nSurvival Rate by Embarkation Port:")
print(train_df.groupby('Embarked')['Survived'].mean() * 100)

# Age vs Survival
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
train_df[train_df['Survived'] == 0]['Age'].hist(bins=30, alpha=0.6, 
                                                  color='red', label='Died')
train_df[train_df['Survived'] == 1]['Age'].hist(bins=30, alpha=0.6, 
                                                  color='green', label='Survived')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution by Survival')
plt.legend()

plt.subplot(1, 2, 2)
sns.boxplot(data=train_df, x='Survived', y='Age', palette='Set2')
plt.title('Age Distribution by Survival (Boxplot)')
plt.xticks([0, 1], ['Died', 'Survived'])

plt.tight_layout()
plt.show()

# Fare vs Survival
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=train_df, x='Survived', y='Fare', palette='Set3')
plt.title('Fare Distribution by Survival')
plt.xticks([0, 1], ['Died', 'Survived'])
plt.ylim(0, 300)

plt.subplot(1, 2, 2)
sns.violinplot(data=train_df, x='Pclass', y='Fare', hue='Survived', 
               palette='muted', split=True)
plt.title('Fare by Class and Survival')
plt.legend(title='Survived', labels=['No', 'Yes'])

plt.tight_layout()
plt.show()
```

```python
# 8. MULTIVARIATE ANALYSIS
# ============================================
print("\n" + "=" * 60)
print("MULTIVARIATE ANALYSIS")
print("=" * 60)

# Create pivot table for Class, Sex, and Survival
pivot_table = train_df.pivot_table(values='Survived', 
                                    index='Sex', 
                                    columns='Pclass', 
                                    aggfunc='mean')
print("\nSurvival Rate by Class and Gender:")
print(pivot_table * 100)

# Visualize
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, fmt='.2%', cmap='RdYlGn', 
            cbar_kws={'label': 'Survival Rate'})
plt.title('Survival Rate by Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Gender')
plt.show()

# Age groups analysis
train_df['Age_Group'] = pd.cut(train_df['Age'], 
                                bins=[0, 12, 18, 35, 60, 100], 
                                labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])

plt.figure(figsize=(12, 6))
sns.countplot(data=train_df, x='Age_Group', hue='Survived', palette='viridis')
plt.title('Survival by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# 9. CORRELATION ANALYSIS
# ============================================
print("\n" + "=" * 60)
print("CORRELATION ANALYSIS")
print("=" * 60)

# Select numerical columns
numerical_cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
correlation_data = train_df[numerical_cols].corr()

print("\nCorrelation Matrix:")
print(correlation_data)

# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', 
            center=0, square=True, linewidths=1)
plt.title('Correlation Heatmap - Titanic Dataset')
plt.tight_layout()
plt.show()

# Pairplot for key variables
plt.figure()
sns.pairplot(train_df[['Survived', 'Pclass', 'Age', 'Fare']], 
             hue='Survived', palette='husl', diag_kind='kde')
plt.suptitle('Pairplot: Key Variables', y=1.02)
plt.show()
```

```python
# 10. OUTLIER DETECTION
# ============================================
print("\n" + "=" * 60)
print("OUTLIER DETECTION")
print("=" * 60)

# Boxplots for numerical variables
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.boxplot(data=train_df, y='Age', ax=axes[0, 0], color='lightblue')
axes[0, 0].set_title('Age - Outlier Detection')

sns.boxplot(data=train_df, y='Fare', ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title('Fare - Outlier Detection')

sns.boxplot(data=train_df, y='SibSp', ax=axes[1, 0], color='lightyellow')
axes[1, 0].set_title('Siblings/Spouses - Outlier Detection')

sns.boxplot(data=train_df, y='Parch', ax=axes[1, 1], color='lightcoral')
axes[1, 1].set_title('Parents/Children - Outlier Detection')

plt.tight_layout()
plt.show()

# Calculate outliers using IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return len(outliers), lower_bound, upper_bound

for col in ['Age', 'Fare', 'SibSp', 'Parch']:
    if col in train_df.columns:
        outlier_count, lower, upper = detect_outliers_iqr(train_df, col)
        print(f"\n{col}:")
        print(f"  Outliers: {outlier_count}")
        print(f"  Lower Bound: {lower:.2f}")
        print(f"  Upper Bound: {upper:.2f}")
```

```python
# 11. FEATURE ENGINEERING INSIGHTS
# ============================================
print("\n" + "=" * 60)
print("FEATURE ENGINEERING INSIGHTS")
print("=" * 60)

# Family Size
train_df['Family_Size'] = train_df['SibSp'] + train_df['Parch'] + 1
print("\nFamily Size Statistics:")
print(train_df['Family_Size'].describe())

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.countplot(data=train_df, x='Family_Size', hue='Survived', palette='Set2')
plt.title('Survival by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])

# Is Alone?
train_df['Is_Alone'] = (train_df['Family_Size'] == 1).astype(int)

plt.subplot(1, 2, 2)
sns.countplot(data=train_df, x='Is_Alone', hue='Survived', palette='Set1')
plt.title('Survival: Alone vs With Family')
plt.xlabel('Traveling Alone (1=Yes, 0=No)')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])

plt.tight_layout()
plt.show()

print("\nSurvival Rate by Family Size:")
print(train_df.groupby('Family_Size')['Survived'].mean() * 100)
```
